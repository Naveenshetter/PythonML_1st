```python
def preprocessor(df):
    res_df=df.copy()
    x=list(df.select_dtypes(include='object').columns)
    for i in x:
        res_df[i]=res_df[i].astype('category')
        res_df[i]=res_df[i].cat.codes
        res_df=pd.get_dummies(res_df,columns=[i],drop_first=True)
    return res_df
```

```python
encoded_df=preprocessor(df)
encoded_df
```
